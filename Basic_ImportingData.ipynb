## Text File

```
filePath = "/content/PlainTextFile.txt"
file = open(file= filePath, mode='r')
text = file.read()
print(file.closed)
file.close()
print(text)
print(file.closed)
```

```
with open(file= filePath, mode='r') as file:
  print(file.read())
print(file.closed)
```
